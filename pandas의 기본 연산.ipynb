```
a = {'이름': ['홍길동', '사람 1', '사람 2'], '나이': [20, 30, 25]}
```

```
df1 = pd.DataFrame(a)
```

```
df1
```

```
b = {'이름': ['홍길동', '사람 1', '사람 2'], '도시': ['서울', '서울', '대전']}
```

```
df2 = pd.DataFrame(b)
```

```
df2
```

```
df = pd.merge(left = df1,
              right = df2,
              how = 'left',
              left_on = '이름',
              right_on = '이름')
```

```
df
```

```
x1 = (df['도시'] == '서울')
```

```
df[x1]
```

```
x2 = (df['도시'] != '서울')
```

```
df[x2]
```

```
x3 = (df['나이'] < 30)
```

```
df[x3]
```

```
x4 = (df['도시'] == '서울') & (df['나이'] < 30)
```

```
df[x4]
```

```
x4 = (df['도시'] == '서울') | (df['나이'] < 30)
```

```
df[x4]
```
